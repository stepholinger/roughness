```python
import geopandas as gpd
import datetime 
from shapely import geometry
import netCDF4 as nc
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import xarray as xr
import rioxarray as rio
import numpy as np
from scipy.fft import rfft,rfftfreq
from scipy import integrate
from scipy.signal import resample
import rasterio
from pyproj import Proj,transform,Geod
from scipy.interpolate import griddata
```

```python
# load landsat image of pig
landsat_scene  = 'LC08_L1GT_001113_20131012_20170429_01_T2'

# read each band
landsat_B4 = rasterio.open('data/imagery/' + landsat_scene+'_B4.TIF') #red
image_B4 = landsat_B4.read(1)

# get coordinates from landsat image
pig_left,pig_bottom,pig_right,pig_top = landsat_B4.bounds
```

```python
# make mask of PIG 
bounds = [pig_left,pig_right,pig_bottom,pig_top]
pig_poly = geometry.Polygon([(bounds[0],bounds[2]),
                    (bounds[0],bounds[3]),
                    (bounds[1],bounds[3]),
                    (bounds[1],bounds[2]),
                    (bounds[0],bounds[2])])
pig_epsg3031 = gpd.GeoDataFrame(geometry=[pig_poly],crs="EPSG:3031")
```

```python
# get PIG ice thicknesses from bedmachine
nc_file = xr.open_dataset('/media/Data/Data/measures/bedmachine/BedMachineAntarctica_2019-11-05_v01.nc')

# extract ice thickness data
thickness = nc_file['thickness']
surface = nc_file['surface']
mask = nc_file['mask']

# get x and y grid
x = nc_file['x']
y = nc_file['y']
xmesh,ymesh = np.meshgrid(x,y)

# set the spatial grid and crs for the data
thickness = thickness.rio.set_spatial_dims('x', 'y')
surface = surface.rio.set_spatial_dims('x', 'y')
mask = mask.rio.set_spatial_dims('x', 'y')
thickness.rio.set_crs("epsg:3031")
surface.rio.set_crs("epsg:3031")
mask.rio.set_crs("epsg:3031")

# clip to pig mask
minx,miny,maxx,maxy = pig_epsg3031.total_bounds
pig_thickness = thickness.rio.clip_box(minx=minx,miny=miny,maxx=maxx,maxy=maxy)
pig_surface = surface.rio.clip_box(minx=minx,miny=miny,maxx=maxx,maxy=maxy)
pig_mask = mask.rio.clip_box(minx=minx,miny=miny,maxx=maxx,maxy=maxy)

# save clipped netCDF file
pig_bedmachine = xr.merge([pig_thickness,pig_surface,pig_mask])
pig_bedmachine.to_netcdf(path="data/thickness/pig_bedmachine.nc")
pig_bedmachine.close()
```

```python
# read in bedmachine thicknesses for just PIG
pig_bedmachine_nc = xr.open_dataset("data/thickness/pig_bedmachine.nc")

# get thickness
thickness_pig = pig_bedmachine_nc['thickness']
elevation_pig = pig_bedmachine_nc['surface']
mask_pig = pig_bedmachine_nc['mask']

# get x and y grid
x_t = pig_bedmachine_nc['x']
y_t = pig_bedmachine_nc['y']

# get grounding line
binary_mask = mask_pig.data
binary_mask[binary_mask == 0] = 0 
binary_mask[binary_mask == 3] = 0 
binary_mask[binary_mask != 0] = 1
k = np.zeros((3,3),dtype=int); k[1] = 1; k[:,1] = 1
grounding_line = binary_dilation(binary_mask==0, k) & binary_mask

# colormap for grounding line
ncolors = 175
color_array = plt.get_cmap('Reds')(range(ncolors))
color_array[:,-1] = np.linspace(0.0,0.75,ncolors)
map_object = LinearSegmentedColormap.from_list(name='gline',colors=color_array)
plt.register_cmap(cmap=map_object)

# plot a specific profile
fig,ax = plt.subplots(2,1,figsize=[10,15])
profile_y = 200
profile_x = [290,360] 
ax[0].plot(thickness_pig['x'][profile_x[0]:profile_x[1]],thickness_pig[profile_y,profile_x[0]:profile_x[1]],'k')

# plot profile on thickness to make sure everything was done correctly
ax[1].pcolormesh(thickness_pig['x'],thickness_pig['y'],thickness_pig,alpha=0.75,cmap="viridis")
ax[1].pcolormesh(mask_pig['x'],mask_pig['y'],grounding_line,cmap="gline")
ax[1].hlines(thickness_pig['y'][profile_y],thickness_pig['x'][profile_x[0]],thickness_pig['x'][profile_x[1]],'C1')
plt.show()
```

```python
def get_roughness(h,d,k_lims):
    diffs = np.diff(d)
    sample_space = np.abs(np.mean(diffs))
    n = len(d)
    k = rfftfreq(n, sample_space)
    yf = sample_space * rfft(h)
    power = np.square(abs(yf))
    psd = power*(k[1]-k[0])
    psd = psd[(k>k_lims[0]) & (k<k_lims[1])]
    k = k[(k>k_lims[0]) & (k<k_lims[1])]
    #plt.loglog(xf,psd)
    int_psd = np.max(integrate.cumulative_trapezoid(psd,xf))
    R = np.sqrt(int_psd)
    return R
```
